# Двоичные числа

Сделайте интерактивную программу, в которой пользователь щелчком мыши может устанавливать или снимать значения нескольких двоичных регистров. Программа должна считать значение получаемого двоичного числа.


## v. 1.0

Двоичная система,

Белый кргу = 0, черный = 1

```python
from tkinter import *
import tkinter
root = tkinter.Tk()

c = Canvas(width=435, height=200, bg='grey90')
c.pack()

sum_t = 0
color = ["white", "black"]
clicks = [0,0,0,0,0,0,0,0]

lbl = Label(root)
lbl.configure(font = ('asinastra',75), anchor=tkinter.W, text=str(sum_t), bg="white", fg="black")
lbl.place(x=0, y=100, height=100, width=440)

oval_1 = c.create_oval(0, 25, 50, 75, fill=color[0], outline = color[0])
oval_2 = c.create_oval(55, 25, 105, 75, fill=color[0], outline = color[0])
oval_3 = c.create_oval(110, 25, 160, 75, fill=color[0], outline = color[0])
oval_4 = c.create_oval(165, 25, 215, 75, fill=color[0], outline = color[0])
oval_5 = c.create_oval(220, 25, 270, 75, fill=color[0], outline = color[0])
oval_6 = c.create_oval(275, 25, 325, 75, fill=color[0], outline = color[0])
oval_7 = c.create_oval(330, 25, 380, 75, fill=color[0], outline = color[0])
oval_8 = c.create_oval(385, 25, 435, 75, fill=color[0], outline = color[0])
fontt = 2000
txt1 = c.create_text(410, 50, font =fontt, text="1", fill=color[1])
txt2 = c.create_text(355, 50, font =fontt, text="2", fill=color[1])
txt3 = c.create_text(300, 50, font =fontt, text="4", fill=color[1])
txt4 = c.create_text(245, 50, font =fontt, text="8", fill=color[1])
txt5 = c.create_text(190, 50, font =fontt, text="16", fill=color[1])
txt6 = c.create_text(135, 50, font =fontt, text="32", fill=color[1])
txt7 = c.create_text(80, 50, font =fontt, text="64", fill=color[1])
txt8 = c.create_text(25, 50, font =fontt, text="128", fill=color[1])

def oval_func_1(event):
    global clicks; global sum_t
    clicks[0] += 1
    if clicks[0] % 2 == 0 and clicks[0] != 0:
        sum_t -= 128
        c.itemconfig(txt8, fill=color[1])
        c.itemconfig(oval_1, fill=color[0], outline=color[0])
        lbl['text'] = sum_t
    elif clicks[0] % 2 != 0 and clicks[0] != 0:
        sum_t += 128
        c.itemconfig(txt8, fill=color[0])
        c.itemconfig(oval_1, fill=color[1], outline=color[1])
        lbl['text'] = sum_t
    
def oval_func_2(event):
    global clicks; global sum_t
    clicks[1] += 1
    if clicks[1] % 2 == 0 and clicks[1] != 0:
        sum_t -= 64
        c.itemconfig(txt7, fill=color[1])
        c.itemconfig(oval_2, fill=color[0], outline=color[0])
        lbl['text'] = sum_t
    elif clicks[1] % 2 != 0 and clicks[1] != 0:
        sum_t += 64
        c.itemconfig(txt7, fill=color[0])
        c.itemconfig(oval_2, fill=color[1], outline=color[1])
        lbl['text'] = sum_t
    
def oval_func_3(event):
    global clicks; global sum_t
    clicks[2] += 1
    if clicks[2] % 2 == 0 and clicks[2] != 0:
        sum_t -= 32
        c.itemconfig(txt6, fill=color[1])
        c.itemconfig(oval_3, fill=color[0], outline=color[0])
        lbl['text'] = sum_t
    elif clicks[2] % 2 != 0 and clicks[2] != 0:
        sum_t += 32
        c.itemconfig(txt6, fill=color[0])
        c.itemconfig(oval_3, fill=color[1], outline=color[1])
        lbl['text'] = sum_t
    
def oval_func_4(event):
    global clicks; global sum_t
    clicks[3] += 1
    if clicks[3] % 2 == 0 and clicks[3] != 0:
        sum_t -= 16
        c.itemconfig(txt5, fill=color[1])
        c.itemconfig(oval_4, fill=color[0], outline=color[0])
        lbl['text'] = sum_t
    elif clicks[3] % 2 != 0 and clicks[3] != 0:
        sum_t += 16
        c.itemconfig(txt5, fill=color[0])
        c.itemconfig(oval_4, fill=color[1], outline=color[1])
        lbl['text'] = sum_t
    
def oval_func_5(event):
    global clicks; global sum_t
    clicks[4] += 1
    if clicks[4] % 2 == 0 and clicks[4] != 0:
        sum_t -= 8
        c.itemconfig(txt4, fill=color[1])
        c.itemconfig(oval_5, fill=color[0], outline=color[0])
        lbl['text'] = sum_t
    elif clicks[4] % 2 != 0 and clicks[4] != 0:
        sum_t += 8
        c.itemconfig(txt4, fill=color[0])
        c.itemconfig(oval_5, fill=color[1], outline=color[1])
        lbl['text'] = sum_t
    
def oval_func_6(event):
    global clicks; global sum_t
    clicks[5] += 1
    if clicks[5] % 2 == 0 and clicks[5] != 0:
        sum_t -= 4
        c.itemconfig(txt3, fill=color[1])
        c.itemconfig(oval_6, fill=color[0], outline=color[0])
        lbl['text'] = sum_t
    elif clicks[5] % 2 != 0 and clicks[5] != 0:
        sum_t += 4
        c.itemconfig(txt3, fill=color[0])
        c.itemconfig(oval_6, fill=color[1], outline=color[1])
        lbl['text'] = sum_t
    
def oval_func_7(event):
    global clicks; global sum_t
    clicks[6] += 1
    if clicks[6] % 2 == 0 and clicks[6] != 0:
        sum_t -= 2
        c.itemconfig(txt2, fill=color[1])
        c.itemconfig(oval_7, fill=color[0], outline=color[0])
        lbl['text'] = sum_t
    elif clicks[6] % 2 != 0 and clicks[6] != 0:
        sum_t += 2
        c.itemconfig(txt2, fill=color[0])
        c.itemconfig(oval_7, fill=color[1], outline=color[1])
        lbl['text'] = sum_t
    
def oval_func_8(event):
    global clicks; global sum_t
    clicks[7] += 1
    if clicks[7] % 2 == 0 and clicks[7] != 0:
        sum_t -= 1
        c.itemconfig(txt1, fill=color[1])
        c.itemconfig(oval_8, fill=color[0], outline=color[0])
        lbl['text'] = sum_t
    elif clicks[7] % 2 != 0 and clicks[7] != 0:
        sum_t += 1
        c.itemconfig(txt1, fill=color[0])
        c.itemconfig(oval_8, fill=color[1], outline=color[1])
        lbl['text'] = sum_t

c.tag_bind(oval_1, '<Button-1>', oval_func_1)
c.tag_bind(txt8, '<Button-1>', oval_func_1)

c.tag_bind(oval_2, '<Button-1>', oval_func_2)
c.tag_bind(txt7, '<Button-1>', oval_func_2)

c.tag_bind(oval_3, '<Button-1>', oval_func_3)
c.tag_bind(txt6, '<Button-1>', oval_func_3)

c.tag_bind(oval_4, '<Button-1>', oval_func_4)
c.tag_bind(txt5, '<Button-1>', oval_func_4)

c.tag_bind(oval_5, '<Button-1>', oval_func_5)
c.tag_bind(txt4, '<Button-1>', oval_func_5)

c.tag_bind(oval_6, '<Button-1>', oval_func_6)
c.tag_bind(txt3, '<Button-1>', oval_func_6)

c.tag_bind(oval_7, '<Button-1>', oval_func_7)
c.tag_bind(txt2, '<Button-1>', oval_func_7)

c.tag_bind(oval_8, '<Button-1>', oval_func_8)
c.tag_bind(txt1, '<Button-1>', oval_func_8)

root.mainloop()
```
